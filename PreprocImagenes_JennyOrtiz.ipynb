```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_height = 256
img_width = 256
batch_size = 32
```

```python
import matplotlib.pyplot as plt
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalizar imágenes
    shear_range=0.2,  # Transformaciones aleatorias
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
validation_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
train_generator = train_datagen.flow_from_directory(
    '/content/dataset/train/',  # Ruta al directorio de entrenamiento
    target_size=(img_height, img_width),  # Redimensionar todas las imágenes
    batch_size=batch_size,
    class_mode='categorical'  # Usar clasificación multiclase
)
```

```python
test_generator = validation_datagen.flow_from_directory(
    '/content/dataset/test/',  # Ruta al directorio de validación
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical'  # Usar clasificación multiclase
)
```

```python
print("Imagenes pre-procesados")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Para visualizar el contenido de las carpetas
# Obtener todas las imágenes y etiquetas
all_images, all_labels = [], []

# Recorrer el dataset completo
for _ in range(len(train_generator)):  # Itera sobre todos los batches
    images, labels = next(train_generator)
    all_images.extend(images)
    all_labels.extend(labels)

# Convertir listas a arrays
all_images = np.array(all_images)
all_labels = np.array(all_labels)

# Mostrar todas las imágenes
fig, axes = plt.subplots(3, 20, figsize=(20, 6))  # 3 filas (una por categoría), 20 columnas
for i, ax in enumerate(axes.flat):
    if i < len(all_images):
        ax.imshow(all_images[i])
        ax.axis('off')
plt.show()

```
