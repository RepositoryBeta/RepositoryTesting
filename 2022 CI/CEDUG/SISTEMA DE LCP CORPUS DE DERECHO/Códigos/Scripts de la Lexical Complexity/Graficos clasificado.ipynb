```python
import pandas as pd
import os
import openpyxl
import matplotlib
import matplotlib.pyplot as plt
```

```python
datos = pd.read_excel('resultados/Metricas_Stanford_ingles.xlsx',engine="openpyxl",sheet_name= 0)
df = pd.DataFrame(datos)

# GRÃFICAS DE LC
plt.figure(figsize=(8,10))
df.groupby('SUBJECT')['LC'].sum().plot(kind='barh',color='#9b9b9b', stacked=True)
plt.title('\nLexical Complexity (LC) Subject\n')
plt.show()

plt.figure(figsize=(7,6))
df.groupby('SEMESTER')['LC'].sum().plot(color='#9b9b9b')
plt.title('\nLexical Complexity (LC) Semester\n')
plt.show()

plt.figure(figsize=(7,8))
df.groupby('CAREER')['LC'].sum().plot(kind='bar',color='#9b9b9b', stacked=True)
plt.title('\nLexical Complexity (LC) Career\n')
plt.show()

# GRÃFICAS DE SSR
plt.figure(figsize=(8,11))
df.groupby('SUBJECT')['SSR'].sum().plot(kind='barh',color='#9b9b9b', stacked=True)
plt.title('\nSPAULDING SPANISH READABILITY (SSR) Subject\n')
plt.show()

plt.figure(figsize=(7,6))
df.groupby('SEMESTER')['SSR'].sum().plot(color='#9b9b9b')
plt.title('\nSPAULDING SPANISH READABILITY (SSR) Semester\n')
plt.show()

plt.figure(figsize=(7,8))
df.groupby('CAREER')['SSR'].sum().plot(kind='bar', color='#9b9b9b',stacked=True)
plt.title('\nSPAULDING SPANISH READABILITY (SSR) Career\n')
plt.show()

# GRÃFICAS DE SCI
plt.figure(figsize=(8,10))
df.groupby('SUBJECT')['SCI'].sum().plot(kind='barh',color='#9b9b9b', stacked=True)
plt.title('\nSENTENCE COMPLEXITY INDEX (SCI) Subject\n')
plt.show()

plt.figure(figsize=(7,6))
df.groupby('SEMESTER')['SCI'].sum().plot(color='#9b9b9b')
plt.title('\nSENTENCE COMPLEXITY INDEX (SCI) Semester\n')
plt.show()

plt.figure(figsize=(7,8))
df.groupby('CAREER')['SCI'].sum().plot(kind='bar',color='#9b9b9b', stacked=True)
plt.title('\nSENTENCE COMPLEXITY INDEX (SCI) Career\n')
plt.show()

# GRÃFICAS DE ARI
plt.figure(figsize=(8,10))
df.groupby('SUBJECT')['ARI'].sum().plot(kind='barh',color='#9b9b9b', stacked=True)
plt.title('\nAUTOMATED READABILITY INDEX (ARI) Subject\n')
plt.show()

plt.figure(figsize=(7,6))
df.groupby('SEMESTER')['ARI'].sum().plot(color='#9b9b9b')
plt.title('\nAUTOMATED READABILITY INDEX (ARI) Semester\n')
plt.show()

plt.figure(figsize=(7,8))
df.groupby('CAREER')['ARI'].sum().plot(kind='bar',color='#9b9b9b', stacked=True)
plt.title('\nAUTOMATED READABILITY INDEX (ARI) Career\n')
plt.show()

# GRÃFICAS DE PM
plt.figure(figsize=(8,11))
df.groupby('SUBJECT')['PM'].sum().plot(kind='barh',color='#9b9b9b', stacked=True)
plt.title('\nPUNCTUATION MARKS (PM) Subject\n')
plt.show()

plt.figure(figsize=(7,6))
df.groupby('SEMESTER')['PM'].sum().plot(color='#9b9b9b')
plt.title('\nPUNCTUATION MARKS (PM) Semester\n')
plt.show()

plt.figure(figsize=(7,8))
df.groupby('CAREER')['PM'].sum().plot(kind='bar',color='#9b9b9b', stacked=True)
plt.title('\nPUNCTUATION MARKS (PM) Career\n')
plt.show()
```
