```python
def definicionTipo(tipo):
  #POS de la oracion
  PROPN="PROPN"  #Nombre Propio
  AUX="AUX"      #Auxiliar
  VERB="VERB"    #Verbo
  ADP="ADP"      #Adposición ejemplo: a, durante
  NOUN="NOUN"    #Sustantivo: niña, gato, Sandra
  NN="NN"        #Nucleo Nominal
  SYM="SYM"      #Simbolo por ejemplo $,%, §, ©, +, -, ×, ÷, =, :), 😝, XD
  NUM="NUM"      #número puede ser 1,2,3
  if tipo==VERB:
      return 0
  elif tipo==NOUN:
      return 1
  elif tipo==PROPN:
      return 2
  elif tipo==ADP:
      return 3
  elif tipo==AUX:
      return 4
  elif tipo==NN:
      return 5
  elif tipo==SYM:
      return 6
  elif tipo==NUM:
      return 7
  else:
      return 8
```

```python
# Funcion para obtener la cantidad de sílabas de una palabra
def getSyllableNumber(token):
  a = pyphen.Pyphen(lang='es')
  word_separada = a.inserted(token)
  print(word_separada)
  if word_separada.find("gano"):
    word_separada = word_separada.replace("gano","ga-no")
  #print("--> word: "+str(word_separada))
  word_separada = word_separada.replace('-',' ')
  word_separada = word_separada.split()
  #return syllables.estimate(token)
  #print(len(word_separada))
  #print(word_separada)
  return len(word_separada)
```

```python
def tokenize(text):
  texto_split = text.split()
  texto = []
  for w in texto_split:
      #s = w.replace(',','').replace(':','').replace(';','').replace('¿','').replace('?','').replace('(','').replace(')','').replace('!','').replace('|','').replace('*','')
      #if(str(s).find("^")):
          #print(s)
      texto.append(w.replace('\'','').replace('"','').replace('¡','').replace(',','').replace(':','').replace(';','').replace('¿','').replace('”','')
      .replace('?','').replace('(','').replace(')','').replace('!','').replace('|','').replace('*','').replace('=','').replace('.-','').replace('“','')
      .replace('.','').replace('+','').replace('{','').replace('}','').replace('[','').replace(']','').replace('<','').replace('#','').replace(';','')
      .replace('$','').replace('%','').replace('\\','').replace('‘','').replace('^','').replace('’','').replace('~','').replace('&','').replace('.(',' '))

      #replace('1)','').replace('2)','')
      #.replace('+','').replace('-','')
  return texto
```

```python
def sintilde(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b)
    return s

print(sintilde("¡Hólá, múndó!"))
print(sintilde("¡HÓLÁ, MÚNDÓ!"))
```
